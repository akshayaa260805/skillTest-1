```python
import numpy as np

# Daily steps for 10 days (example)
steps = np.array([5000, 7200, 8000, 6500, 9000, 10000, 7500, 8200, 6000, 9500])

# Statistics
total_steps = np.sum(steps)
avg_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print("Total Steps:", total_steps)
print("Average Steps:", avg_steps)
print("Maximum Steps:", max_steps)
print("Minimum Steps:", min_steps)

# High activity days
threshold = 8000
high_days = steps[steps > threshold]

print("High Activity Days (steps > 8000):", high_days)
```
