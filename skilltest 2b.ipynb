```python
import numpy as np

# Marks of students (Math, Science, English)
marks = np.array([
    [78, 85, 80],
    [65, 70, 75],
    [90, 88, 92],
    [55, 60, 58],
    [82, 79, 85]
])

# Subject-wise statistics
print("Average Marks:", np.mean(marks, axis=0))
print("Highest Marks:", np.max(marks, axis=0))
print("Lowest Marks:", np.min(marks, axis=0))

# Total marks of each student
total = np.sum(marks, axis=1)

# Top performing student
top_student = np.argmax(total)

print("Top Student Index:", top_student)
print("Top Student Total Marks:", total[top_student])
```
