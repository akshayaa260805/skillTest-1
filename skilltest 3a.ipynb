```python
import numpy as np

# Impedance values
Z1 = 6 + 3j
Z2 = 2 - 2j

# Series impedance
Z_series = Z1 + Z2

# Parallel impedance
Z_parallel = (Z1 * Z2) / (Z1 + Z2)

print("Impedance 1:", Z1)
print("Impedance 2:", Z2)
print("Total Series Impedance:", Z_series)
print("Total Parallel Impedance:", Z_parallel)
```
