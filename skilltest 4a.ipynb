```python
import pandas as pd

# Example EEG data
data = {
    "Participant": ["P1","P1","P1","P2","P2","P2"],
    "Electrode": ["E1","E2","E3","E1","E2","E3"],
    "Signal": [12.5, 14.2, 13.8, 11.9, 15.1, 14.0]
}

df = pd.DataFrame(data)

# Statistical analysis
mean_value = df["Signal"].mean()
variance_value = df["Signal"].var()
peak_value = df["Signal"].max()

print("Mean:", mean_value)
print("Variance:", variance_value)
print("Peak Value:", peak_value)
```
