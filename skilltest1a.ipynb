```python
scores = [65, 78, 45, 90, 55, 62, 88, 49, 73, 60]

print("Average:", sum(scores)/len(scores))
print("Highest:", max(scores))
print("Lowest:", min(scores))

for s in scores:
    if s >= 50:
        result = "Pass"
    else:
        result = "Fail"

    if s >= 85:
        grade = "A"
    elif s >= 70:
        grade = "B"
    elif s >= 60:
        grade = "C"
    elif s >= 50:
        grade = "D"
    else:
        grade = "F"

    print(s, result, grade)
```
