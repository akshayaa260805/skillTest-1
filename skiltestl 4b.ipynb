```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
'Product': ['Laptop','Smartphone','Tablet','Monitor','Keyboard'],
'Price': [1200,800,300,150,50],
'Quantity': [10,20,15,25,50]
}

df = pd.DataFrame(data)

# Basic operations
print("Inventory Data:")
print(df)

print("Average Price:", df['Price'].mean())
print("Maximum Price:", df['Price'].max())
print("Minimum Quantity:", df['Quantity'].min())

# Total inventory value
df['Total Value'] = df['Price'] * df['Quantity']
print("Total Inventory Value:", df['Total Value'].sum())

# Plot
plt.bar(df['Product'], df['Price'])
plt.title("Product Price Distribution")
plt.xlabel("Products")
plt.ylabel("Price")
plt.show()
```
